```python
#Read in the park_info_with_2022_visitor_info (this is a csv file that contains both the NPS scraped data and the NPS API call data)
# (i.e. the monthly visitors for 2022 and the park information)
visitor_info_df = pd.read_csv('../Output/park_info_with_2022_visitor_info.csv')
visitor_info_df
```

```python
#Create an empty list that will hold all information about parks
parks_list = []

# Loop through each row of the DataFrame
for index, row in visitor_info_df.iterrows():
    
    #Initialize an empty dictionary that will store this row's information
    data_dict = {}
    
    # Extract the values from the dataframe and store them as variables
    name = row['Park']
    fullName = row['Full Park Name']
    state = row['State']
    monthlyVisitors = [row['Jan-22'], row['Feb-22'], row['Mar-22'], row['Apr-22'], row['May-22'], row['Jun-22'],
                      row['Jul-22'], row['Aug-22'], row['Sep-22'], row['Oct-22'], row['Nov-22'], row['Dec-22']]
    totalVisitors2022 = row['totalVisitors2022']
    designation = row['Designation']
    entranceFee = row['entranceFee']
    entranceFeeDesc = row['entranceFeeDesc']
    
    # Add each of these variables to the dictionary
    data_dict["name"] = name
    data_dict["fullParkName"] = fullName
    data_dict["state"] = state
    data_dict["monthlyVisitors"] = monthlyVisitors
    data_dict["totalVisitors2022"] = totalVisitors2022
    data_dict["designation"] = designation
    data_dict["entranceFee"] = entranceFee
    data_dict["entranceFeeDesc"] = entranceFeeDesc
    
    #Append the dictionary to the parks_list
    parks_list.append(data_dict)
    
parks_list
```

```python
#This function removes any states where the state is more than two letters (to get rid of lists of states)

all_parks_list = []

#It runs through all the parks list
for i in range(len(parks_list)):
    
    data_dict = {}

    #If the state is a string or the length of the state is two or fewer words
    if not isinstance(parks_list[i]["state"], str) or len(parks_list[i]["state"]) <= 2:
        
        #Then basically keep everything the same
        data_dict["name"] = parks_list[i]["name"]
        data_dict["fullParkName"] = parks_list[i]["fullParkName"]
        data_dict["state"] = parks_list[i]["state"]
        data_dict["monthlyVisitors"] = parks_list[i]["monthlyVisitors"]
        data_dict["totalVisitors2022"] = parks_list[i]["totalVisitors2022"]
        data_dict["designation"] = parks_list[i]["designation"]
        data_dict["entranceFee"] = parks_list[i]["entranceFee"]
        data_dict["entranceFeeDesc"] = parks_list[i]["entranceFeeDesc"]
        
        
    #Otherwise (if the state is longer than two letters, or is nan)
    else:
        
        #Keep everything else the same but replace state with " "
        data_dict["name"] = parks_list[i]["name"]
        data_dict["fullParkName"] = parks_list[i]["fullParkName"]
        data_dict["state"] = " "
        data_dict["monthlyVisitors"] = parks_list[i]["monthlyVisitors"]
        data_dict["totalVisitors2022"] = parks_list[i]["totalVisitors2022"]
        data_dict["designation"] = parks_list[i]["designation"]
        data_dict["entranceFee"] = parks_list[i]["entranceFee"]
        data_dict["entranceFeeDesc"] = parks_list[i]["entranceFeeDesc"]

    #Append this dictionary to the all_parks_list.
    #all_parks_list will be the master list that contains all park info (including Yelp reviews by the end)
    all_parks_list.append(data_dict)
    
all_parks_list
```
