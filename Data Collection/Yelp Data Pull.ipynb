```python
visitor_info_df = pd.read_csv('../Output/park_info.csv')
visitor_info_df
```

```python
parks_list = []

# Loop through each row of the DataFrame
for index, row in visitor_info_df.iterrows():
    
    data_dict = {}
    
    # Extract the values of 'name' and 'state'
    name = row['Name']
    state = row['State']
    
    # Add the values to the dictionary
    data_dict["Park Name"] = name
    data_dict["State"] = state
    
    parks_list.append(data_dict)
    
parks_list
```

```python
# Remove parks that have more than 1 state associated
filtered_parks_list = []

for i in range(len(parks_list)):
    
    data_dict = {}
        
    if len(parks_list[i]["State"]) <= 2:
        
        data_dict["Park Name"] = parks_list[i]["Park Name"]
        data_dict["State"] = parks_list[i]["State"]
    
    else:
        data_dict["Park Name"] = parks_list[i]["Park Name"]
        data_dict["State"] = " "

    filtered_parks_list.append(data_dict)
    
filtered_parks_list
```
