## Preparing Data for Monthly Visitor Plot

```python
#Import dependences
import pandas as pd
import numpy as np
import re
import json
```

```python
park_info = pd.read_csv('../Resources/Monthly Visitors By Park.csv')
park_info
```

```python
park_info = park_info[["Park",
                     "Jan-22",
                    "Feb-22",
                    "Mar-22",
                    "Apr-22",
                    "May-22",
                     "Jun-22",
                     "Jul-22",
                     "Aug-22",
                     "Sep-22",
                    "Oct-22",
                    "Nov-22",
                    "Dec-22"]]
park_info
```

```python
#Convert all monthly visitor columns to numbers
park_info["Jan-22"] = park_info["Jan-22"].str.replace(',', '').apply(pd.to_numeric)
park_info["Feb-22"] = park_info["Feb-22"].str.replace(',', '').apply(pd.to_numeric)
park_info["Mar-22"] = park_info["Mar-22"].str.replace(',', '').apply(pd.to_numeric)
park_info["Apr-22"] = park_info["Apr-22"].str.replace(',', '').apply(pd.to_numeric)
park_info["May-22"] = park_info["May-22"].str.replace(',', '').apply(pd.to_numeric)
park_info["Jun-22"] = park_info["Jun-22"].str.replace(',', '').apply(pd.to_numeric)
park_info["Jul-22"] = park_info["Jul-22"].str.replace(',', '').apply(pd.to_numeric)
park_info["Aug-22"] = park_info["Aug-22"].str.replace(',', '').apply(pd.to_numeric)
park_info["Sep-22"] = park_info["Sep-22"].str.replace(',', '').apply(pd.to_numeric)
park_info["Oct-22"] = park_info["Oct-22"].str.replace(',', '').apply(pd.to_numeric)
park_info["Nov-22"] = park_info["Nov-22"].str.replace(',', '').apply(pd.to_numeric)
park_info["Dec-22"] = park_info["Dec-22"].str.replace(',', '').apply(pd.to_numeric)

park_info
```

```python
#Change NaNs to 0s

columns = ["Jan-22",
            "Feb-22",
            "Mar-22",
            "Apr-22",
            "May-22",
             "Jun-22",
             "Jul-22",
             "Aug-22",
             "Sep-22",
            "Oct-22",
            "Nov-22",
            "Dec-22"]

for column in columns:
    park_info[column] = park_info[column].fillna(0)

park_info
```

```python
#Get totalVisitors2022 column
park_info["totalVisitors2022"] = park_info["Jan-22"] + park_info["Feb-22"]  + park_info["Mar-22"] + park_info["Apr-22"] + park_info["May-22"] + park_info["Jun-22"] + park_info["Jul-22"] + park_info["Aug-22"] + park_info["Sep-22"] + park_info["Oct-22"] + park_info["Nov-22"] + park_info["Dec-22"]
park_info
```
