## Preparing Data for Monthly Visitor Plot

```python
#Import dependences
import pandas as pd
import numpy as np
import re
import json
```

```python
park_info = pd.read_csv('../Output/park_info_with_2022_visitor_info.csv')
park_info
```

```python
yelp_info = pd.read_csv('../Output/yelp_info.csv')
yelp_info
```

```python
yelp_info = yelp_info.rename(columns = {"Name": "Park"})
yelp_info
```

```python
all_graph_info = park_info.merge(yelp_info, on = "Park")
all_graph_info
```

```python
all_graph_info = all_graph_info[["Park",
                               "Full Park Name_x",
                                "Designation",
                                "Park Code",
                               "Jan-22", 
                               "Feb-22", 
                               "Mar-22", 
                               "Apr-22", 
                               "May-22", 
                               "Jun-22", 
                               "Jul-22", 
                               "Aug-22", 
                               "Sep-22",  
                               "Oct-22",  
                               "Nov-22",  
                               "Dec-22", 
                               "totalVisitors2022",
                               "entranceFee",
                               "entranceFeeDesc",
                               "Yelp Rating"]]

all_graph_info
```

```python
all_graph_info = all_graph_info.rename(columns = {"Full Park Name_x": "Full Park Name"})
all_graph_info
```

```python
graph_data = []

for i, row in all_graph_info.iterrows():
    
    #Initialize an empty dictionary that will store this row's information
    park_dict = {}
    
    # Extract the values from the dataframe and store them as variables
    parkName = row['Park']
    fullName = row['Full Park Name']
    parkDesignation = row['Designation']
    parkCode = row['Park Code']
    totalVisitors2022 = row['totalVisitors2022']
    yelpRating = row['Yelp Rating']
    entranceFee = row['entranceFee']
    
    monthlyVisitors = [row["Jan-22"],
                      row["Feb-22"],
                      row["Mar-22"],
                      row["Apr-22"],
                      row["May-22"],
                      row["Jun-22"],
                      row["Jul-22"],
                      row["Aug-22"],
                      row["Sep-22"],
                      row["Oct-22"],
                      row["Nov-22"],
                      row["Dec-22"]]
    
    
    # Add each of these variables to the dictionary
    park_dict["parkName"] = parkName
    park_dict["fullName"] = fullName
    park_dict["parkDesignation"] = parkDesignation 
    park_dict["parkCode"] = parkCode
    park_dict["monthlyVisitors"] = monthlyVisitors
    park_dict["totalVisitors2022"] = totalVisitors2022
    park_dict["yelpRating"] = yelpRating
    park_dict["entranceFee"] = entranceFee

    graph_data.append(park_dict)
    
graph_data
```

```python
#Write the dictionary to a json file
with open("../Output/graph_data.json", "w") as outfile:
    json.dump(graph_data, outfile)
```
