```python
#Import dependences
import pandas as pd
import numpy as np
import re
import json
```

```python
#Read last year's monthly visitor data into a Pandas DataFrame

jan_2022_df = pd.read_csv('../Resources/Monthly Visitors By Park/Jan Visitors By Park.csv')
feb_2022_df = pd.read_csv('../Resources/Monthly Visitors By Park/Feb Visitors By Park.csv')
mar_2022_df = pd.read_csv('../Resources/Monthly Visitors By Park/Mar Visitors By Park.csv')
apr_2022_df = pd.read_csv('../Resources/Monthly Visitors By Park/Apr Visitors By Park.csv')
may_2022_df = pd.read_csv('../Resources/Monthly Visitors By Park/May Visitors By Park.csv')
jun_2022_df = pd.read_csv('../Resources/Monthly Visitors By Park/Jun Visitors By Park.csv')
jul_2022_df = pd.read_csv('../Resources/Monthly Visitors By Park/Jul Visitors By Park.csv')
aug_2022_df = pd.read_csv('../Resources/Monthly Visitors By Park/Aug Visitors By Park.csv')
sep_2022_df = pd.read_csv('../Resources/Monthly Visitors By Park/Sep Visitors By Park.csv')
oct_2022_df = pd.read_csv('../Resources/Monthly Visitors By Park/Oct Visitors By Park.csv')
nov_2022_df = pd.read_csv('../Resources/Monthly Visitors By Park/Nov Visitors By Park.csv')
dec_2022_df = pd.read_csv('../Resources/Monthly Visitors By Park/Dec Visitors By Park.csv')
```

```python
#Merge all the dataframes together

monthly_visitors_2022_df = jan_2022_df.merge(feb_2022_df, on = "Park")
monthly_visitors_2022_df = monthly_visitors_2022_df.merge(mar_2022_df, on = "Park")
monthly_visitors_2022_df = monthly_visitors_2022_df.merge(apr_2022_df, on = "Park")
monthly_visitors_2022_df = monthly_visitors_2022_df.merge(may_2022_df, on = "Park")
monthly_visitors_2022_df = monthly_visitors_2022_df.merge(jun_2022_df, on = "Park")
monthly_visitors_2022_df = monthly_visitors_2022_df.merge(jul_2022_df, on = "Park")
monthly_visitors_2022_df = monthly_visitors_2022_df.merge(aug_2022_df, on = "Park")
monthly_visitors_2022_df = monthly_visitors_2022_df.merge(sep_2022_df, on = "Park")
monthly_visitors_2022_df = monthly_visitors_2022_df.merge(oct_2022_df, on = "Park")
monthly_visitors_2022_df = monthly_visitors_2022_df.merge(nov_2022_df, on = "Park")
monthly_visitors_2022_df = monthly_visitors_2022_df.merge(dec_2022_df, on = "Park")
monthly_visitors_2022_df
```

```python
#Get rid of the unnecessary columns
monthly_visitors_2022_df = monthly_visitors_2022_df[["Park",
                                                     "Jan-22",
                                                    "Feb-22",
                                                    "Mar-22",
                                                    "Apr-22",
                                                    "May-22",
                                                     "Jun-22",
                                                     "Jul-22",
                                                     "Aug-22",
                                                     "Sep-22",
                                                    "Oct-22",
                                                    "Nov-22",
                                                    "Dec-22"]]
monthly_visitors_2022_df
```

```python
#Convert all monthly visitor columns to numbers
monthly_visitors_2022_df["Jan-22"] = monthly_visitors_2022_df["Jan-22"].str.replace(',', '').apply(pd.to_numeric)
monthly_visitors_2022_df["Feb-22"] = monthly_visitors_2022_df["Feb-22"].str.replace(',', '').apply(pd.to_numeric)
monthly_visitors_2022_df["Mar-22"] = monthly_visitors_2022_df["Mar-22"].str.replace(',', '').apply(pd.to_numeric)
monthly_visitors_2022_df["Apr-22"] = monthly_visitors_2022_df["Apr-22"].str.replace(',', '').apply(pd.to_numeric)
monthly_visitors_2022_df["May-22"] = monthly_visitors_2022_df["May-22"].str.replace(',', '').apply(pd.to_numeric)
monthly_visitors_2022_df["Jun-22"] = monthly_visitors_2022_df["Jun-22"].str.replace(',', '').apply(pd.to_numeric)
monthly_visitors_2022_df["Jul-22"] = monthly_visitors_2022_df["Jul-22"].str.replace(',', '').apply(pd.to_numeric)
monthly_visitors_2022_df["Aug-22"] = monthly_visitors_2022_df["Aug-22"].str.replace(',', '').apply(pd.to_numeric)
monthly_visitors_2022_df["Sep-22"] = monthly_visitors_2022_df["Sep-22"].str.replace(',', '').apply(pd.to_numeric)
monthly_visitors_2022_df["Oct-22"] = monthly_visitors_2022_df["Oct-22"].str.replace(',', '').apply(pd.to_numeric)
monthly_visitors_2022_df["Nov-22"] = monthly_visitors_2022_df["Nov-22"].str.replace(',', '').apply(pd.to_numeric)
monthly_visitors_2022_df["Dec-22"] = monthly_visitors_2022_df["Dec-22"].str.replace(',', '').apply(pd.to_numeric)

monthly_visitors_2022_df
```

```python
#Get totalVisitors2022 column
monthly_visitors_2022_df["totalVisitors2022"] = monthly_visitors_2022_df["Jan-22"] + monthly_visitors_2022_df["Feb-22"]  + monthly_visitors_2022_df["Mar-22"] + monthly_visitors_2022_df["Apr-22"] + monthly_visitors_2022_df["May-22"] + monthly_visitors_2022_df["Jun-22"] + monthly_visitors_2022_df["Jul-22"] + monthly_visitors_2022_df["Aug-22"] + monthly_visitors_2022_df["Sep-22"] + monthly_visitors_2022_df["Oct-22"] + monthly_visitors_2022_df["Nov-22"] + monthly_visitors_2022_df["Dec-22"]
monthly_visitors_2022_df
```

```python
#Pull all park names

park_names = monthly_visitors_2022_df["Park"].unique()
park_names = park_names.tolist()
park_names
```

```python
#Use regex to get all the initials (NM, NP, etc.)

#Create an empty array for the initials
initials_list = []

pattern = '\s([A-Z\s&]+)$'

for item in park_names:
    matches = re.findall(pattern, item)
    if matches:
        initials_list.append(matches[-1])

print(initials_list)


```

```python
#Create a list of unique park initials
unique_initials_list = []

for item in initials_list:
    if item not in unique_initials_list:
        unique_initials_list.append(item)

unique_initials_list       
```

```python
unique_initials_list
```

```python
#Manually create the dictionary that matches each initial to a name
initial_names_dict = {
    'NHP & Ecological Pres' : "National Historical Park and Ecological Preserve",
    'HQ NHS' : 'Headquarters National Historic Site',
    'NM & PRES': 'National Monument & Preserve',
    'NP & PRES': 'National Park & Preserve',
    'NM & HS': 'National Monument and Historic Shrine',
    'MEM PKWY': 'Memorial Parkway',
    'NHP & PRES' : 'National Park & Preserve',
    'NPRES': 'National Preserve',
    'NMEM': 'National Memorial',
    'NRRA': 'National River',
    'PKWY': 'National Parkway',
    'NRES': 'National Reserve',
    'W&SR': 'Wild & Scenic River',
    'S&RR': 'Scenic & Recreational River',
    'NHP': 'National Historical Park',
    'NMP': 'National Military Park',
    'NHS': 'National Historic Site', 
    'NRA': 'National Recreation Area',
    'MEM':'Memorial',
    'NBP': 'National Battlefield Park',
    'NRR': 'National Recreational River',
    'NSR' : 'National Scenic River',
    'EHP': 'Ecological & Historic Preserve',
    'NM': 'National Monument',
    'NB': 'National Battlefield',
    'NL': 'National Lakeshore',
    'NP': 'National Park',
    'NS': 'National Seashore',
    'NR': 'National River',
    'HS': 'Historic Shrine',
    'PRES': 'Preserve'
}
```

```python
len(unique_initials_list)
```

```python
#Go through the park names and replace each initial with its corresponding name using regex
for i in range(len(park_names)):
        for initial, full_name in initial_names_dict.items():
            pattern = r'\b' + re.escape(initial) + r'(?=\s|&|$)'
            park_names[i] = re.sub(pattern, full_name, park_names[i])

park_names
```

```python
#Create a column with the full name
monthly_visitors_2022_df.insert(1, "Full Park Name", park_names)
monthly_visitors_2022_df
```

```python
# Read in all the park information

park_info_df = pd.read_csv('../Output/park_info_NPS_API.csv')
park_info_df
```

```python
#Rename columns for merging and for js inputting to map

park_info_df = park_info_df.rename(columns = {"Name": "Full Park Name",
                                              "Park Code": "parkCode",
                                              "Park ID": "parkID",
                                             "Entrance Fee": "entranceFee",
                                             "Entrance Fee Desc": "entranceFeeDesc"})
park_info_df
```

```python
#Merge the two dataframes to contain all data (including any NaNs)
all_df = monthly_visitors_2022_df.merge(park_info_df, on = "Full Park Name", how = "outer")
all_df
```

```python
all_df.to_csv("../Output/all_park_info.csv", encoding='utf8', index=False)
```

```python
#Merge the two dataframes to only contain ones with Lat and Long information (a left merge in this case)

merged_df = monthly_visitors_2022_df.merge(park_info_df, on = "Full Park Name", how = "left")
merged_df
```

```python
merged_df = merged_df.dropna(subset=["Latitude"])
merged_df = merged_df.dropna(subset=["Longitude"])
merged_df
```
